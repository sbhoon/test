```python
data = pd.read_csv("score.csv")
```

```python
data['total'] = data.eng + data.kor + data.math + data.sci
```

```python
data = data[['eng', 'kor', 'math', 'sci']]
```

```python
np.sum(data.values, axis=1)
```

```python
data = data[data.columns[:-1]]
data
```

```python
data['avg'] = data.total / 4
```

```python
data['best_subject'] = data.columns[np.argmax(data.iloc[:,:4].values,axis=1)]
```

```python
data['greade'] = np.where(data['avg'] >= 90, 'A',
np.where(data['avg'] >= 80, 'B', 
np.where(data['avg'] >= 70, 'C', 'D')))
```
